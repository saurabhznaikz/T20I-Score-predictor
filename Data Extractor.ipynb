```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filenames = []
for file in os.listdir('data'):
    filenames.append(os.path.join('data',file))
```

```python
filenames[0:5]
```

```python
final_df = pd.DataFrame()
counter = 1
for file in tqdm(filenames):
    with open(file, 'r') as f:
        df = pd.json_normalize(safe_load(f))
        df['match_id'] = counter
        final_df = final_df.append(df)
        counter+=1
        
```

## Extracting ball by ball data for 1st innings score

```python
count = 1
delivery_df = pd.DataFrame()
for index, row in final_df.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_of_dismissed,
            'city':city,
            'venue':venue
        })
    delivery_df = delivery_df.append(loop_df)
```

```python
delivery_df
```

## Need bowling team as a input for our model so dereiving our column by our dataset

```python
#function to find bowling team
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl,axis=1)
delivery_df
```

## Finding and handling missing values

```python
output.isnull().sum()
```

## Here we can see we can replace missing city values by venues name's first word as it represents location only

```python
output['city']=np.where(output['city'].isnull(),output['venue'].str.split().apply(lambda x:x[0]),output['city'])
```

```python
output.isnull().sum()
```

## Null values are handles using venue column and we can drop venue now as it is not needed required ahead

```python
output.drop(columns=['venue'],inplace=True)
output
```

```python
output['city'].value_counts()
```

## As city is a categorical column and we need to convert it into numerical features later, we will only consider cities having played >= 5 matches i.e 600 balls

```python
eligible_city = output['city'].value_counts()[output['city'].value_counts()>600].index.tolist()
```

```python
output=output[output['city'].isin(eligible_city)]
```

```python
output
```

## Derieveing current score using cumalative fuction on runs

```python
output['current_score']=output.groupby('match_id').cumsum()['runs']
```

```python
output.head(10)
```

## Derieving balls bowled and ball left column

```python
output['over'] = output['ball'].apply(lambda x:str(x).split(".")[0])
output['ball_no'] = output['ball'].apply(lambda x:str(x).split(".")[1])
output
```

```python
output['balls_bowled'] = (output['over'].astype('int')*6) + output['ball_no'].astype('int')
output
```

```python
output['balls_left'] = 120 - output['balls_bowled']
output['balls_left'] = output['balls_left'].apply(lambda x:0 if x<0 else x)
output
```

## derieving wickets left by using player dismissed column

```python
output['player_dismissed'] = output['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
output['player_dismissed'] = output['player_dismissed'].astype('int')
output['player_dismissed'] = output.groupby('match_id').cumsum()['player_dismissed']
output['wickets_left'] = 10 - output['player_dismissed']
```

```python
output
```

## derieving current runrate from current score

```python
output['crr'] = (output['current_score']*6)/output['balls_bowled']
```

```python
output
```

```python
groups = output.groupby('match_id')

match_ids = output['match_id'].unique()
last_three = []
for id in match_ids:
    last_three.extend(groups.get_group(id).rolling(window=18).sum()['runs'].values.tolist())
```

```python
output['last_three'] = last_three
```

```python
output
```

## Adding total runs column

```python
final_df =output.groupby('match_id').sum()['runs'].reset_index().merge(output,on='match_id')
final_df.head()
```

## Shuffling before train test split

```python
final_df = final_df.sample(final_df.shape[0])
final_df
```

```python
## Train Test Split
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

## Using one hot encoding for column transformation

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

## Creating a pipeline for encoding,standardization and model creation

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

## Fitting train data into the pipeline and then test the test data using r2 score

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
